```python
import pandas as pd
movie_dataset = pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
movie_dataset
```

```python
# Task 1
num_rows, num_columns = movie_dataset.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
# Display the titles and genres of the first 50 movies
first_50_movies = movie_dataset.head(50)
print(first_50_movies[['title', 'genres']])
   
```

```python
# Task 2 
# Check for null values in each column
null_columns = movie_dataset.isnull().sum()
print("Columns with null values:")
print(null_columns[null_columns > 0])
movie_dataset['budget'].fillna(movie_dataset['budget'].mean(), inplace=True)
movie_dataset['genres'].fillna("Unknown", inplace=True)
movie_dataset.dropna(inplace=True)
    
```

```python
# Task 3
# Filter the dataset to select rows where the budget is greater than $220,000
budget_threshold = 220000
filtered_movies = movie_dataset[movie_dataset['budget'] > budget_threshold]

# Extract and display the genres of the filtered movies
selected_genres = filtered_movies['genres']
print("Movie categories (genres) with a budget greater than $220,000:")
print(selected_genres.unique())
```

```python
# Task 4 
# Filter the dataset to select rows where the revenue is greater than $961,000,000
revenue_threshold = 961000000
filtered_movies = movie_dataset[movie_dataset['revenue'] > revenue_threshold]

# Extract and display the genres of the filtered movies
selected_genres = filtered_movies['genres']
print("Movie categories (genres) with revenue greater than $961,000,000:")
print(selected_genres.unique())
```

```python
# Task 5 
# Remove rows with budget and revenue equal to 0
movie_dataset = movie_dataset[(movie_dataset['budget'] != 0) & (movie_dataset['revenue'] != 0)]
```

```python
#Task 6a
# Sort the dataset by 'revenue' column in descending order
top_10_revenue_movies = movie_dataset.sort_values(by='revenue', ascending=False).head(10)

# Display the top 10 movies with the highest revenues
print("Top 10 Movies with Highest Revenues:")
print(top_10_revenue_movies[['title', 'revenue']])
```

```python
# Task 6b
# Sort the dataset by 'budget' column in ascending order
top_10_budget_movies = movie_dataset.sort_values(by='budget').head(10)

# Display the top 10 movies with the least budget
print("\nTop 10 Movies with Least Budget:")
print(top_10_budget_movies[['title', 'budget']])
```

```python
# Task 7
# Calculate the correlation between 'budget' and 'popularity'
correlation = movie_dataset['budget'].corr(movie_dataset['popularity'])

print(f"Correlation between budget and popularity: {correlation:.2f}")
```

```python
# Task 9
# Split the 'production_companies' column into a list of production company names
production_companies_list = movie_dataset['production_companies'].str.split('|')

# Flatten the list of production company names
flat_production_companies_list = [company for companies in production_companies_list for company in companies]

# Count the occurrences of each production company
company_count = pd.Series(flat_production_companies_list).value_counts()

# Display the top 25 production companies in descending order
top_25_companies = company_count.head(25)
print("Top 25 Production Companies by Number of Movies Produced:")
print(top_25_companies)
```

```python
# Task 10 
# Sort the dataset by 'revenue' column in descending order and select the top 500 movies
top_500_revenue_movies = movie_dataset.sort_values(by='revenue', ascending=False).head(500)
# Calculate mean, median, and mode for 'budget' column
budget_mean = top_500_revenue_movies['budget'].mean()
budget_median = top_500_revenue_movies['budget'].median()
budget_mode = top_500_revenue_movies['budget'].mode().values[0]

# Calculate mean, median, and mode for 'revenue' column
revenue_mean = top_500_revenue_movies['revenue'].mean()
revenue_median = top_500_revenue_movies['revenue'].median()
revenue_mode = top_500_revenue_movies['revenue'].mode().values[0]

# Calculate mean, median, and mode for 'runtime' column
runtime_mean = top_500_revenue_movies['runtime'].mean()
runtime_median = top_500_revenue_movies['runtime'].median()
runtime_mode = top_500_revenue_movies['runtime'].mode().values[0]

# Display the measures of central tendency
print("Measures of Central Tendency for Budget:")
print(f"Mean: {budget_mean:.2f}")
print(f"Median: {budget_median:.2f}")
print(f"Mode: {budget_mode:.2f}")

print("\nMeasures of Central Tendency for Revenue:")
print(f"Mean: {revenue_mean:.2f}")
print(f"Median: {revenue_median:.2f}")
print(f"Mode: {revenue_mode:.2f}")

print("\nMeasures of Central Tendency for Runtime:")
print(f"Mean: {runtime_mean:.2f}")
print(f"Median: {runtime_median:.2f}")
print(f"Mode: {runtime_mode:.2f}")
import matplotlib.pyplot as plt

# Create subplots for box plots
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.boxplot(top_500_revenue_movies['budget'], vert=False)
plt.title('Budget')

plt.subplot(1, 3, 2)
plt.boxplot(top_500_revenue_movies['revenue'], vert=False)
plt.title('Revenue')

plt.subplot(1, 3, 3)
plt.boxplot(top_500_revenue_movies['runtime'], vert=False)
plt.title('Runtime')

plt.tight_layout()
plt.show()

```

```python
# Task 11
# Calculate the average runtime for the top 500 revenue movies
average_runtime = top_500_revenue_movies['runtime'].mean()

# Filter the movies with runtimes above the average
above_average_runtime_movies = top_500_revenue_movies[top_500_revenue_movies['runtime'] > average_runtime]

# Display the names of the movies along with their run times
print("Movies with Above Average Runtime:")
print(above_average_runtime_movies[['title', 'runtime']])

```
